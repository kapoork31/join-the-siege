```python
import requests
```

```python
# Set the URL of your FastAPI endpoint
url = 'http://127.0.0.1:8000/uploadfile/'

# Define the customer_id and the file to be uploaded
customer_id = 123
file_path = 'new_invoice.txt'  # Update the file path accordingly

# Open the file to send as multipart form-data
with open(file_path, 'rb') as file:
    # Send the POST request
    response = requests.post(
        url,
        params={'customer_id': customer_id},
        files={'file': (file_path, file, 'application/pdf')}  # File upload
    )

# Check the response
if response.status_code == 200:
    print('File uploaded successfully!')
    print(response.json())  # Print the response from the API
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

```python
# Define the URL and the JSON payload (body of the POST request)
url = 'http://127.0.0.1:8000/classify_file'
data = {
    'filename': 'new_invoice.txt',
    'customer_id': 123
}

# Send the POST request with the JSON body
response = requests.post(url, json=data)

# Check the response
if response.status_code == 200:
    print("Success!")
    print(response.json())  # Prints the returned JSON response
else:
    print(f"Error {response.status_code}: {response.text}")
```

```python
import re
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline, FunctionTransformer
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
# Function to clean text by lowercasing, removing punctuation, and removing stop words
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Remove stop words
    text = " ".join([word for word in text.split() if word not in ENGLISH_STOP_WORDS])
    
    # Remove extra whitespace
    text = " ".join(text.split())
    
    return text
```

```python

documents =  driver_licenses + bank_statements + invoices
processed_documents = [clean_text(doc) for doc in documents]
```

```python
labels = ['driver_license'] * len(driver_licenses) + ['bank_statement'] * len(bank_statements) + ['invoice'] * len(invoices)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(processed_documents, labels, test_size=0.3, random_state=42)

# Create a CountVectorizer with unigrams, bigrams, and trigrams
vectorizer = CountVectorizer(ngram_range=(1, 3), stop_words='english')

# Create a pipeline with a text preprocessing step and a classifier
model = make_pipeline(
    CountVectorizer(ngram_range=(1, 3), stop_words='english'),  # Vectorizer with unigrams, bigrams, trigrams
    RandomForestClassifier(n_estimators=100, random_state=42)
)
```

```python
# Add the preprocessing step to the pipeline
 
# Fit the model
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))

# To classify new documents
def classify_file(text):
    return model.predict([text])[0]

# Example usage:
new_file = "Some raw text representing a bank statement"
file_type = classify_file(new_file)
print(f"The file type is classified as: {file_type}")
```

```python
import pickle

# Define the file path for the serialized model
model_filename = 'text_classifier_pipeline.pkl'

# Save the entire pipeline
with open(model_filename, 'wb') as file:
    pickle.dump(model, file)

print(f"Pipeline saved to {model_filename}")
```
